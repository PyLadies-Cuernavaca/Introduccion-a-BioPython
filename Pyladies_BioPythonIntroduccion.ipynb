# **Introducción a Biopython**

Biopython es una librería de Python que contiene métodos para procesar y analizar con facilidad datos biológicos [1] [2].


Aunque su funcionalidad es variada, es muy útil para procesar la información contenida en secuencias de ADN y PROTEÍNAS, así como para procesar información relacionada con la secuencias.


Algunos de los métodos y clases disponbibles son:

   *  Objetos **Seq and SeqRecord** creados para el manejo de secuencias.

        - **Seq**: Utiliza una serie de métodos para manejar las secuencias como **strings de Python**

        - **SeqRecord**: Formado por librerías útiles para **inspeccionar y extraer** información de los registros asociados a las secuencias tomando encuenta **distintos formatos**.
   
   *  **Bio.SeqIO** - sequence input/output.

      - Como su nombre lo indica proporciona una interfase uniforme para dar lectura, escribir y dar salida a archivos de secuencias en distintos formatos.
      
      - Puede usar datos provenientes de bases de datos diversas.

   *  **Bio.AlignIO** - alignment input/output.

      - Permite el manejo y formateo de archivos derivados de distintos métodos de alineaminto.

   *  **Bio.PopGen** - population genetics

      - Analisis poblacionales que nos provee de métodos estadísticos usales en genética de poblaciones.

   *  **Bio.PDB** - structural bioinformatic.

    - Módulo enfocado en el análisis de estructuras cristalográficas de macromoléculas biológicas.

   *  **Biopython’s** BioSQL interface

   - Es un esfuerzo conjunto entre los proyectos OBF (BioPerl, BioJava, etc.) para respaldar un esquema de base de datos compartido para almacenar datos de secuencia.


 Todas estas librerías y sus métodos puede descargar desde https://biopython.org/.

#Objetos mutables:

Al igual que la cadena tipo (string), el objeto Seq es de "solo lectura", o en la terminología de Python **inmutable**. Por lo que transformalo en un **objeto modificable o mutable** es útil.

```python
#Alternativamente, puedes crear un objeto mutable directamente desde una cadena
mutable_seq = MutableSeq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
mutable_seq
#type(mutable_seq)
```

```python
print ("Nueva longitud",len(mutable_seq))
```

Podemos aplicar metodos como reverse a esta secuencia mutable

# **Bio.SeqRecord**: El objeto SeqRecord almacena descripciones o metadatos asociados a las secuencias genómicas

Este módulo te permite trabajar con los metadatos asociados a secuencias descargadas de diversas bases de datos

**Atributos principales:**

**id** - Identificador, como una etiqueta de locus (cadena)

**seq** - La secuencia en sí (objeto Seq o similar)

**Atributos adicionales:**

**name** - Nombre de la secuencia, p. ej., nombre del gen (cadena)

**description** - Texto adicional (cadena)

**dbxrefs** - Lista de referencias cruzadas de la base de datos (lista de cadenas)

**features** - Cualquier (sub)feature definida (lista de objetos SeqFeature)

**annotations** - Información adicional sobre toda la secuencia (diccionario). La mayoría de las entradas son cadenas o listas de cadenas.

**letter_annotations** - Anotación por letra/símbolo (diccionario restringido). Contiene secuencias de Python (listas, cadenas o tuplas) cuya longitud coincide con la de la secuencia. Un uso típico sería contener una lista de números enteros que representan puntuaciones de calidad de secuenciación, o una cadena que representa la estructura secundaria.

```python
from Bio.SeqRecord import SeqRecord
#help(SeqRecord)
```

Usar un objeto SeqRecord no es muy complicado, ya que toda la información se presenta como atributos de la clase. Por lo general, como usuario de base de datos no  crearas un SeqRecord “a mano”, sino que utilizará Bio.SeqIO para leer un archivo de secuencia. Sin embargo, crear SeqRecord puede ser bastante sencillo.

```python
#Creando un "SeqRecord"
from Bio.Seq import Seq

#asignamos la secuencia a una variable
simple_seq = Seq("GATC")

#importamos la librería
from Bio.SeqRecord import SeqRecord

#ahora la convertimos en un objeto SeqRecord
simple_seq_r = SeqRecord(simple_seq)

#creemos un identificador para mi secuencia
simple_seq_r.id = "AC12345"

#hagammos una descripción
simple_seq_r.description = "Mi primer secuencia en SeqRecord"
print(simple_seq_r.id,simple_seq_r.description)
simple_seq_r.seq
```

EL objeto Bio.SeqFeauture tiene los siguientes atributos

1. **location** - the location of the feature on the sequence (FeatureLocation)

2. **type** - the specified type of the feature (ie. CDS, exon, repeat…)

3. **location_operator** - a string specifying how this SeqFeature may be related to others.

4. **strand** - A value specifying on which strand (of a DNA sequence, for instance) the feature deals with. 1 indicates the plus strand, -1 indicates the minus strand, 0 indicates stranded but unknown (? in GFF3), while the default of None indicates that strand doesn’t apply (dot in GFF3, e.g. features on proteins). Note this is a shortcut for accessing the strand property of the feature’s location.

5. **id** - A string identifier for the feature.

6. **ref** - A reference to another sequence. This could be an accession number for some different sequence. Note this is a shortcut for the reference property of the feature’s location.

7. **ref_db** - A different database for the reference accession number. Note this is a shortcut for the reference property of the location

8. **qualifiers** - A dictionary of qualifiers on the feature. These are analogous to the qualifiers from a GenBank feature table. The keys of the dictionary are qualifier names, the values are the qualifier values. As of Biopython 1.69 this is an ordered dictionary.

###Secuenciación masiva: ejercicio

El resultado de la secuenciaciación de alto rendimiento se expresa en archivos **fastq**. Que a diferencia del formato fasta tienen además **los valores de calidad de cada nucleótido secuenciado, representados en formato PHRED (PHRED qualiy --código ascii--)**. A continuación realizaremos un gráfico para analizar estas calidades:

Originalmente fue desarrollado para el llamado de base de Phred ( Phred base calling, en Inglés) que es un programa de cómputo diseñado para la identificación de una secuencia de bases a partir de señales de fluorescencia generados por un secuenciador de ADN automatizado

Base-Calling of Automated Sequencer Traces UsingPhred. I. Accuracy Assessment https://genome.cshlp.org/content/8/3/175

El nivel de calidad Phred o Q, se define como la relación logarítmica que guarda la probabilidad de tener errores en el llamado de bases (P)

               Q = -10log10 PoP = 10 -Q/10

Un ejemplo prácitco

              Q = -10log10 PoP = 10 30/10

En donde hemos tomado una base en concreto con un nivel de  calidad de 30: Lo que significa que las probabilidades de que esta base sea incorrecta es de 1 entre 1000. Lo que implica es que tenemos un 99.9% de problabilidad de error.

Costruyamos un gráfico para evaluar las calidades de nuestras secuencias.

 En este ejemplo, extraeremos la calidad PHRED promedio por posición en una lectura, así como los valores de calidad altos intermedios y bajos.

Para después generar un gráfico que muestre las calidades organizadas de acuerdo las clasificaciones establecidas.





```python
#importando librerías
from Bio import SeqIO
import pylab

archivo = open("/content/drive/MyDrive/Pyladies/Sample_soil_1.fastq")
parse = list(SeqIO.parse(archivo,"fastq"))
tam_seq = []
max_seqs = []
min_seqs = []
media = []

for i in parse:
	tam_seq.append(len(i.seq))

max_seq = max(tam_seq) #tamaño máximo de secuencia
min_seq = min(tam_seq) #tamaño mínimo de secuencia
total_seq = len(parse) #longitud de  secuencias

for j in range(max_seq):
	aux = 0
	mas = 0
	mis = max_seq
	for i in parse:

		if(len(i.seq) < max_seq):
			i.letter_annotations\
["phred_quality"].append(0) #pegar cada núclótido anotado
		aux = aux + \
i.letter_annotations["phred_quality"][j] #almacenar cada nucleótido en la varible auxiliar y acumulamos cada valor
#para cada posición

 #tomando los valores de cada nucleótido anotado
		if mas < i.letter_annotations\
["phred_quality"][j]:
			mas = i.letter_annotations\
["phred_quality"][j]

		if mis > i.letter_annotations\
["phred_quality"][j]:
			mis = i.letter_annotations\
["phred_quality"][j]
 #acumulando la calidad por posición
	media.append(int(aux/total_seq))
	max_seqs.append(mas)
	min_seqs.append(mis)

pylab.plot(media,'-k',label="Media", linewidth=2) #dibuja una linea negra (-k),de ancho 2
pylab.plot(max_seqs,'-g',label="Max") #dibuja una linea de color verde
pylab.plot(min_seqs,'-r',label="Min") #dibuja una linea de color rojo
pylab.grid(color='gray', linestyle=':', linewidth=1) #cuadricula
pylab.legend() #Crendo el recuadro con la leyenda
pylab.ylim(0,45)
pylab.ylabel("PHRED") #título del eje "y"
pylab.xlabel("Lectura (pb)") #título del eje "x"
pylab.savefig("/content/drive/MyDrive/biophy_data/Gráfico_calidades_sample1.png") #salvando archivo

pylab.show()
```
